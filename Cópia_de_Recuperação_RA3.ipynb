#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(-2, 2.1, 0.1)
x1 = (y**2)-2
x2 = y
plt.fill_between(y, x1, x2, where=(x2 > x1), color='blue', alpha=0.5)

plt.plot(y, x1, label='x1=y^2-2')
plt.plot(y, x2, label='x2=y')
plt.title('Gráfico')
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 6 (Vale até 2 pontos)

a) Calcule a integral indefinida:
$$\int (3x^2 + 2e^x) \, dx$$

b) Calcule a integral indefinida:
$$\int x \sin(x) \, dx$$

c) Calcule a integral definida:
$$\int_0^{\pi} x \cos(x) \, dx$$

d) Calcule a integral indefinida:
$$\int x^2 e^x \, dx$$


## Respostas


$$a)$$
$$\int \:3x^2+2e^xdx$$
$$=\int \:3x^2dx+\int \:2e^xdx$$
$$\int \:3x^2dx=x^3$$
$$\int \:2e^xdx=2e^x$$
$$=x^3+2e^x$$

$$b) $$
$$\int \:x\sin \left(x\right)dx$$

$$=-x\cos \left(x\right)-\int \:-\cos \left(x\right)dx$$

$$\int \:-\cos \left(x\right)dx=-\sin \left(x\right) $$

$$=-x\cos \left(x\right)-\left(-\sin \left(x\right)\right) $$

$$=-x\cos \left(x\right)+\sin \left(x\right) $$


$$c)$$
$$\int_0^{\pi} x \cos(x) \, dx$$

$$=\left[x\sin \left(x\right)-\int \sin \left(x\right)dx\right]_0^{\pi }$$

$$\int \sin \left(x\right)dx=-\cos \left(x\right)$$

$$=\left[x\sin \left(x\right)-\left(-\cos \left(x\right)\right)\right]_0^{\pi }$$

$$=\left[x\sin \left(x\right)+\cos \left(x\right)\right]_0^{\pi }$$

$$=-2$$

$$d)$$
$$\int x^2 e^x \, dx$$

$$=x^2e^x-\int \:2xe^xdx$$

$$\int \:2xe^xdx=2\left(e^xx-e^x\right)$$

$$=x^2e^x-2\left(e^xx-e^x\right)$$
