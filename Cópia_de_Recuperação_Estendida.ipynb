#Questão 7 (vale 1 ponto)
Considere um fenômeno físico em que a posição de um objeto em movimento varia com o tempo. A taxa de variação da posição em relação ao tempo possui um componente trigonométrico e é dada pela função $f(t) = A * cos(t)$, onde $A$ é uma constante que determina a amplitude da variação. Sua tarefa é plotar o gráfico da função da taxa de variação e da função posição deste objeto considerando que $A$ é um número inteiro que varia entre 2 e 5.

## Resposta

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


t = sp.symbols('t')
A = sp.symbols('A')

fx = (fx, A* (sp.cos(t)))

f = sp.lambdify(t, A, fx, "numpy")
x_vals = np.linspace(2, 5, 1000)
f_vals = f(x_vals)


plt.plot(x_vals, f_vals)
plt.title('Grafico')
plt.legend()
plt.show()
```
