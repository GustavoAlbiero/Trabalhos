```python
#1
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
    return x**3 + 5*x**2 - 7*x + 9

def f_prime(x):
    return 3*x**2 + 10*x - 7

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values for f(x) and f'(x)
y = f(x)
y_prime = f_prime(x)

# Plot f(x) and f'(x)
plt.plot(x, y, label='f(x)')
plt.plot(x, y_prime, label="f'(x)")

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x)
def g(x):
    return np.exp(x) * np.cos(x)

# Define the derivative of g(x)
def g_prime(x):
    return np.exp(x) * (np.cos(x) - np.sin(x))

# Generate x values
x = np.linspace(-5, 5, 1000)

# Calculate y values for g'(x)
y_prime = g_prime(x)

# Plot the derivative graph
plt.plot(x, y_prime)
plt.xlabel('x')
plt.ylabel("g'(x)")
plt.title("Derivada g(x) = e^x * cos(x)")
plt.grid(True)
plt.show()
```

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
import numpy as np

def f(x):
    return x**2 - 98*x + 4


x = np.linspace(0, 100, 1000)
y = f(x)
max_value = np.max(y)
max_x = x[np.argmax(y)]
print("Ponto máximo de f(x) em 0 <= x <= 100:", max_value)
```
