A seguir estão listados três limites. Sua tarefa será encontar o valor destes limites de duas formas: algébricamente e gráficamente.

b) Para solucionar graficamente deverá usar uma tabela de aproximação pela esquerda e pela direita. Lembre-se de plotar o gráfico, os pontos usados para a aproximação e de imprimir a tabela no terminal.

2. $$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1}$$

1. <br>
$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$ <br><br>
$\lim _{x\to \:0}\left(\frac{2\sin \left(2x\right)}{2x}\right)$ <br><br>
$\lim _{x\to \:0}\left(\frac{4\cos \left(2x\right)}{2}\right)$ <br><br>
$\frac{4\cos \left(2\cdot \:0\right)}{2}$ <br><br>
$=2$


```python
#1.
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    return (1 - np.cos(2 * x)) / (x ** 2)

# Generate x values from -1 to 1 (excluding 0) with small increments
x = np.linspace(-1, 1, 1000)
x = x[x != 0]

# Calculate the corresponding function values
y = function(x)

# Plot the graph
plt.plot(x, y, label='f(x) = (1 - cos(2x)) / x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('1.   Graph of f(x) = (1 - cos(2x)) / x^2')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

2. <br>
$\lim _{x\to \infty \:}\left(\frac{3x^2-2x+1}{2x^2+x-1}\right)$ <br>
<br>
$\lim _{x\to \infty \:}\left(\frac{3-\frac{2}{x}+\frac{1}{x^2}}{2+\frac{1}{x}-\frac{1}{x^2}}\right)$<br>
<br>
$\frac{\lim _{x\to \infty \:}\left(3-\frac{2}{x}+\frac{1}{x^2}\right)}{\lim _{x\to \infty \:}\left(2+\frac{1}{x}-\frac{1}{x^2}\right)}$ <br>
<br>
$=\frac{3}{2}$

```python
#2.
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    return (3 * x ** 2 - 2 * x + 1) / (2 * x ** 2 + x - 1)

# Generate x values from -10 to 10 with small increments
x = np.linspace(-10, 10, 1000)

# Calculate the corresponding function values
y = function(x)

# Plot the graph
plt.plot(x, y, label='f(x) = (3x^2 - 2x + 1) / (2x^2 + x - 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('2.   Graph of f(x) = (3x^2 - 2x + 1) / (2x^2 + x - 1)')

# Add legend
plt.legend()

# Display the plot
plt.show()

```

3. <br>
$\lim _{x\to \:0}\left(x\cdot \ln \left(1+\sin \left(\frac{1}{2}\right)\right)\right)$ <br><br>
$\lim _{x\to \:0}\left(x\ln \left(1+\sin \left(\frac{1}{2}\right)\right)\right)$ <br><br>
$0\cdot \ln \left(1+\sin \left(\frac{1}{2}\right)\right)$<br><br>
$=0$


```python
#3.
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    return x * np.log(1 + np.sin(1 / x))

# Generate x values from -1 to 1 excluding 0, with small increments
x = np.linspace(-1, 1, 1000)
x = x[x != 0]

# Calculate the corresponding function values
y = function(x)

# Plot the graph
plt.plot(x, y, label='f(x) = x * ln(1 + sin(1/x))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('3.   Graph of f(x) = x * ln(1 + sin(1/x))')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
